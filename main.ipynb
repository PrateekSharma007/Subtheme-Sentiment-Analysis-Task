```python
import re
from collections import Counter
def count_sentence_frequency(column):
    # Convert the column to a list of sentences (strings)
    sentences = column.astype(str).tolist()
    # Count the frequency of each sentence
    sentence_counts = Counter(sentences)
    return sentence_counts



frequency_data = [] #in this we wil store like i said above
for col in df.columns:
    word_counts = count_sentence_frequency(df[col])
    for word, count in word_counts.items():
       frequency_data.append({'Column': col, 'Sentence': word, 'Frequency': count}) #we make a dataframe for the frequency 

frequency_df = pd.DataFrame(frequency_data) #we make a dataframe for the frequency 
frequency_df = frequency_df.sort_values(by='Frequency', ascending=False)
```

```python
df_melted = df.melt(id_vars=['Sentence'], value_vars=['Subtheme1', 'Subtheme2', 'Subtheme3'], var_name='SubthemeType', value_name='Subtheme')
df_melted = df_melted.dropna(subset=['Subtheme'])
df_one_hot = pd.get_dummies(df_melted, columns=['Subtheme'])
df_one_hot = df_one_hot.groupby('Sentence').sum().reset_index()
print(df_one_hot)
# print(df_one_hot)
```

```python
import pandas as pd

def dataModification(data):
    data1 = data.iloc[:, 1:]  # Assuming the first column contains the sentences
    
    # Create a DataFrame with index and columns from frequency_data
    data2 = pd.DataFrame(index=range(0, data1.shape[0]), columns=data1.stack().unique())
    
    # Iterate over each row in data1
    for i, row in data1.iterrows():
        # Get a list of subthemes in the current row, dropping NaNs
        l_subthemes = row.dropna().tolist()
        
        # Set the corresponding columns to 1
        for t in l_subthemes:
            data2.loc[i, t] = 1
            
    # Fill NaNs with 0s
    data2 = data2.fillna(0)
    
    # Combine the original sentences with the one-hot encoded subthemes
    final_data = pd.DataFrame()
    final_data['Sentence'] = data.iloc[:, 0]  # Assuming the first column contains the sentences
    for column in data2.columns:
        final_data[column] = data2[column]
        
    return final_data
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
# ps.stem("I am checking Stemming for data science internship ")
```

```python
#This is function which I have used in my earlier project to remove the stopwords

import string
def transform_text(text) : 
    text = text.lower()
    text = nltk.word_tokenize(text)  #the text has come in the list
    y=[]
    for i in text: 
        if i.isalnum(): #special character are removed
            y.append(i)

    text = y[:] 
    y.clear() 
    for i in text : 
        if i not in stopwords.words('english') and i not in string.punctuation : 
            y.append(i)

    text = y[:]
    y.clear()
    for i in text : 
        y.append(ps.stem(i)) 



    return " ".join(y)
```
